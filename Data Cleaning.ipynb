```python
df['Fullname'] = df['First'] + ' ' + df['Last']
```

```python
df['Time'] = pd.to_timedelta(df['Time'])
```

```python
df['Total']=df['Time'].dt.total_seconds() / 60
```

```python
df['Time']
```

```python
df['Total'] = df['Total'].round().astype(int)
```

```python
df.rename(columns={'Total' : 'Total_Minutes'})
```

```python
df.rename(columns={'Division' : 'Gender'})
```

```python
from geopy.geocoders import Nominatim
```

```python
nom=Nominatim(user_agent="running")
```

```python
df["Coordinate"] = df["City"].apply(nom.geocode)
```

```python
df["Coordinate"]
```

```python
df["latitude"]=df["Coordinate"].apply(lambda x: x.latitude if x !=None else None)
```

```python
df["longitude"] = df["Coordinate"].apply(lambda x: x.longitude if x !=None else None)
```

```python
df = df.drop(columns=["Coordinate"])
```
